```python
quarter_1,quarter_2 = np.percentile(stroke_df["Avg_Glucose_Level"],[25,75])
inter_quarter_range = quarter_2 - quarter_1
lower_fence = quarter_1 - (1.5 * inter_quarter_range)
upper_fence = quarter_2 + (1.5 * inter_quarter_range)

sort_outliers = stroke_df[(stroke_df["Avg_Glucose_Level"] > lower_fence) & (stroke_df["Avg_Glucose_Level"] < upper_fence)]


q_1,q_2 = np.percentile(sort_outliers["Bmi"],[25,75])
iqr = q_2 - q_1
lower_fence = q_1 - (1.5 * iqr)
upper_fence = q_2 + (1.5 * iqr)

outliers_df = sort_outliers[(sort_outliers["Bmi"] > lower_fence) & (sort_outliers["Bmi"] < upper_fence)]


age_1,age_2 = np.percentile(outliers_df["Age"],[25,75])
age_iqr = age_2 - age_1
lower_fence = age_1 - (1.5 * age_iqr)
upper_fence = age_2 + (1.5 * age_iqr)

outliers_age = outliers_df[(outliers_df["Age"] > lower_fence) & (outliers_df["Age"] < upper_fence)]


hyper_11,hyper_22 = np.percentile(outliers_age["Avg_Glucose_Level"],[25,75])
hyper_iqr = hyper_22 - hyper_11
lower_fence = hyper_11 - (1.5 * hyper_iqr)
upper_fence = hyper_22 + (1.5 * hyper_iqr)

outliers_hyper = outliers_age[(outliers_age["Avg_Glucose_Level"] > lower_fence) & (outliers_age["Avg_Glucose_Level"] < upper_fence)]

hyper_3,hyper_4 = np.percentile(outliers_hyper["Avg_Glucose_Level"],[25,75])
hyper1_iqr = hyper_4 - hyper_3
lower_fence = hyper_3 - (1.5 * hyper1_iqr)
upper_fence = hyper_4 + (1.5 * hyper1_iqr)

outliers_hyper1 = outliers_hyper[(outliers_hyper["Avg_Glucose_Level"] > lower_fence) & (outliers_hyper["Avg_Glucose_Level"] < upper_fence)]

hyper_5,hyper_6 = np.percentile(outliers_hyper1["Avg_Glucose_Level"],[25,75])
hyper2_iqr = hyper_6 - hyper_5
lower_fence = hyper_5 - (1.5 * hyper2_iqr)
upper_fence = hyper_6 + (1.5 * hyper2_iqr)

outliers_hyper2 = outliers_hyper1[(outliers_hyper1["Avg_Glucose_Level"] > lower_fence) & (outliers_hyper1["Avg_Glucose_Level"] < upper_fence)]

hyper_7,hyper_8 = np.percentile(outliers_hyper2["Avg_Glucose_Level"],[25,75])
hyper3_iqr = hyper_8 - hyper_7
lower_fence = hyper_7 - (1.5 * hyper3_iqr)
upper_fence = hyper_8 + (1.5 * hyper3_iqr)

outliers_hyper3 = outliers_hyper2[(outliers_hyper2["Avg_Glucose_Level"] > lower_fence) & (outliers_hyper2["Avg_Glucose_Level"] < upper_fence)]

hypert_1,hypert_2 = np.percentile(outliers_hyper3["Bmi"],[25,75])
hypert_iqr = hypert_2 - hypert_1
lower_fence = hypert_1 - (1.5 * hypert_iqr)
upper_fence = hypert_2 + (1.5 * hypert_iqr)

hypert_df = outliers_hyper3[(outliers_hyper3["Bmi"] > lower_fence) & (outliers_hyper3["Bmi"] < upper_fence)]

hyperten_1,hyperten_2 = np.percentile(hypert_df["Work_Type"],[25,75])
hyperten_iqr = hyperten_2 - hyperten_1
lower_fence = hyperten_1 - (1.5 * hyperten_iqr)
upper_fence = hyperten_2 + (1.5 * hyperten_iqr)

hyperten_df = hypert_df[(hypert_df["Work_Type"] > lower_fence) & (hypert_df["Work_Type"] < upper_fence)]


plt.figure(figsize=(20,7))
sns.boxplot(hyperten_df);
```

```python
# Split the dataset
X = hyperten_df.iloc[:,:-1]
Y = hyperten_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

trainx,testx,trainy,testy = train_test_split(X,Y,test_size = 0.3,random_state = 2)

scale = StandardScaler()

x_tain = scale.fit_transform(trainx)
x_tast = scale.transform(testx)
print(x_tain.shape,x_tast.shape)
print(trainy.shape,testy.shape)
```

```python
import tensorflow as tf

# initialize random seed
tf.random.set_seed(42)

# create model
modal = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10,activation = "relu"),
    tf.keras.layers.Dense(10,activation = "relu"),
    tf.keras.layers.Dense(10,activation = "relu"),
    tf.keras.layers.Dense(10,activation = "relu"),
    tf.keras.layers.Dense(1,activation = "sigmoid"),
    
])

modal.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)

lr = tf.keras.callbacks.LearningRateScheduler(lambda epochs: 1e-2 * 10 **(epochs/20))

modele = modal.fit(x_tain,trainy,epochs = 200,callbacks= [lr],verbose = 0)
```
